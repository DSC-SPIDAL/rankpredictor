```python
# calc the section rank
sections=['S1','S2','S3','S4','S5']
final_lap = max(rankdata.last_lap)

def calc_abs_rank_delta(l1,l2):
    """
    l1,l2 is the input of sorted car numbers
    returns the differences of the rank in the two lists
    """
    if len(l1)!=len(l2):
        print('rank_delta calc error:l1 len=%d,l2 len=%d'%(len(l1),len(l2)))
        return -1
    
    #assume that they are the same set
    s1 = np.sort(l1)
    s2 = np.sort(l2)
    if np.sum(s1 - s2) != 0:
        print('rank_delta calc error:l1 and l2 are different cars')
        return -2
    
    #argsort indices are the ranks
    r1 = np.argsort(l1)
    r2 = np.argsort(l2)
    dist = np.sum(np.abs(r1-r2))
    return dist
    
    
def make_section_rankdb(rankdata):
    final_lap = max(rankdata.last_lap)
    l_prev = []
    # lap, section, dist
    distarry = np.zeros((final_lap,len(sections)))
    for lap in range(1,final_lap+1):
        for idx, section in enumerate(sections):
            #get l
            l = list(rankdata[(rankdata['last_lap']==lap) & 
                              (rankdata['section_identifier']==section)].car_number.values)
            
            if l_prev:
                dist = calc_abs_rank_delta(l_prev, l)
                if dist < 0:
                    print('lap=%d, section=%s'%(lap, section))
                    print('l1 cars=%s'%(','.join([str(x) for x in l_prev])))
                    print('l2 cars=%s'%(','.join([str(x) for x in l])))
                distarry[lap-1, idx] = dist
                
            #update 
            l_prev = l
    
    return distarry

def check_sections(rankdata, carno):
    """
    output the section map
    """
    rankdata = rankdata[rankdata['car_number']==carno]
    
    final_lap = max(rankdata.last_lap)
    # lap, section, dist
    distarry = np.zeros((final_lap,len(sections)))
    for lap in range(1,final_lap+1):
        for idx, section in enumerate(sections):
            #get l
            l = list(rankdata[(rankdata['last_lap']==lap) & 
                              (rankdata['section_identifier']==section)].car_number.values)
            
            if len(l) > 0:
                distarry[lap-1, idx] = 1
    
    return distarry
```
